### Project 2: Employee Performance Analysis
#### Objective: Merge employee details with performance review data and calculate the average performance score department.

```python
import pandas as pd
```

##### Step 1: Create Employee Details DataFrame

```python
employees = pd.DataFrame({
    'employee_id': [101, 102, 103, 104, 105],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Ella'],
    'department': ['HR', 'IT', 'IT', 'Finance', 'HR']
})

print("Employee Details:")
print(employees)
```

##### Step 2: Create Performance Review DataFrame

```python
performance = pd.DataFrame({
    'employee_id': [101, 102, 103, 104, 105],
    'performance_score': [85, 90, 78, 88, 92]
})

print("\nPerformance Review Data:")
print(performance)
```

##### Step 3: Merge Both DataFrames on 'employee_id'

```python
merged_df = pd.merge(employees, performance, on='employee_id', how='left')

print("\nMerged Employee Performance Data:")
print(merged_df)

# how = left is a left join, employees is left table and performance is the right table it will join from left table
```

##### Step 4: Calculate Average Performance per Department

```python
avg_performance = merged_df.groupby('department')['performance_score'].mean().reset_index()

print("\nAverage Performance Score by Department:")
print(avg_performance)
```
